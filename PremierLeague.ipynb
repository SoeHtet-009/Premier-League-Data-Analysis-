```python
data = pd.read_csv('EPL_standings_2000-2022.csv')
data.head()
```

```python
data.info()
```

```python
# Maximum number of wins in a season
data[data['W']==max(data['W'])]
```

```python
# Maximum number of losses in a season
data[data['L']==max(data['L'])]
```

```python
# Maximum number of draws in a season
data[data['D']==max(data['D'])]
```

```python
# Premier league winners over the last 22 years
winners = data[data['Pos']==1]
winners
```

```python
# Which team has won premier league titles before?
winners['Team'].unique()
```

```python
winners['Team'].max()
```

```python
winner_counts = winners['Team'].value_counts()
winner_counts
```

```python
winner_counts.plot.bar()
plt.show()
```

```python
# Who won with more than or equal to 100 pts?
winners[winners.Pts >= 100]
```

```python
# Boxplot of premier league's winners total points
winners.boxplot("Pts")
```

```python
relegation = data[data["Pos"] == 18]
relegation
```

```python
relegation['Pts'].mean()
```

```python
# How many points on average are needed to avoid relegation?
safeteam = data[data["Pos"] == 17]

safepoint = safeteam['Pts'].mean()
safepoint
```

```python
# How many points on average are needed to secure champions league football
UCL = data[data["Pos"] == 4]
UCL['Pts'].mean()
```

```python
# How did chelsea perform over the last 22 years?
Chelsea = data[data["Team"]=="Chelsea"]
plt.bar(Chelsea["Season"],Chelsea['Pts'])
plt.xticks(Chelsea["Season"], rotation='vertical')
plt.title("Chelsea's Performance")
plt.xlabel("Season")
plt.ylabel("Total Points")
plt.show()
```

```python
# How is Chelsea's total goals scored trend over the years?
plt.bar(Chelsea["Season"],Chelsea['GF'])
plt.xticks(Chelsea["Season"], rotation='vertical')
plt.title("Chelsea's Goals Scored")
plt.xlabel("Season")
plt.ylabel("Goals Scored")
plt.show()
```

```python
# How is Chelsea's total goals conceded trend over the years?
plt.bar(Chelsea["Season"],Chelsea['GA'])
plt.xticks(Chelsea["Season"], rotation='vertical')
plt.title("Chelsea's Goals Conceded")
plt.xlabel("Season")
plt.ylabel("Goals Conceded")
plt.show()
```

```python
# Chelsea best goal scoring season
Chelsea[Chelsea["GF"]==Chelsea["GF"].max()]
```

```python
# Chelsea best defence season
Chelsea[Chelsea["GA"]==Chelsea["GA"].min()]
```

```python
# Chelsea best goal difference season
Chelsea[Chelsea["GD"]==Chelsea["GD"].max()]
```

```python
# Rise of Man City 
ManCity = data[data["Team"]=="Manchester City"]
plt.plot(ManCity["Season"],ManCity['Pts'])
plt.xticks(rotation='vertical')
plt.show()
```
